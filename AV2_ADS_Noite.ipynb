```python
import os
import sys

def sc_isalpha(c: str) -> bool:
    if not c or len(c) != 1: return False
    return ('a' <= c <= 'z') or ('A' <= c <= 'Z')

def sc_isdigit(c: str) -> bool:
    if not c or len(c) != 1: return False
    return '0' <= c <= '9'

def sc_isalnum(c: str) -> bool:
    return sc_isalpha(c) or sc_isdigit(c)

def sc_isascii(c: str) -> bool:
    if not c or len(c) != 1: return False
    return 0 <= ord(c) <= 127

def sc_isprint(c: str) -> bool:
    if not c or len(c) != 1: return False
    return 32 <= ord(c) <= 126

def sc_toupper(c: str) -> str:
    if 'a' <= c <= 'z':
        return chr(ord(c) - 32)
    return c

def sc_tolower(c: str) -> str:
    if 'A' <= c <= 'Z':
        return chr(ord(c) + 32)
    return c

def sc_atoi(s: str) -> int:
    if not s: return 0
    i = 0
    sign = 1
    res = 0

    while i < len(s) and s[i] in ' \t\n\r\v\f':
        i += 1

    if i < len(s) and (s[i] == '-' or s[i] == '+'):
        if s[i] == '-':
            sign = -1
        i += 1

    while i < len(s) and sc_isdigit(s[i]):
        res = res * 10 + (ord(s[i]) - ord('0'))
        i += 1

    return res * sign

```

```python
def sc_memset(b: list, c: str, length: int) -> list:
    for i in range(min(length, len(b))):
        b[i] = c
    return b

def sc_bzero(b: list, length: int) -> None:
    sc_memset(b, '\0', length)

def sc_memcpy(dst: list, src: list, length: int) -> list:
    if not dst and not src: return dst
    for i in range(min(length, len(src))):
        if i < len(dst):
            dst[i] = src[i]
        else:
            dst.append(src[i])
    return dst

def sc_memmove(dst: list, src: list, length: int) -> list:
    if not dst and not src: return dst
    if dst > src:
        for i in range(min(length, len(src)) - 1, -1, -1):
            if i < len(dst): dst[i] = src[i]
    else:
        sc_memcpy(dst, src, length)
    return dst

def sc_memchr(s: list, c: str, length: int):
    for i in range(min(length, len(s))):
        if s[i] == c:
            return s[i:]
    return None

def sc_memcmp(s1: list, s2: list, length: int) -> int:
    for i in range(length):
        if i >= len(s1) and i >= len(s2): return 0
        if i >= len(s1): return -ord(s2[i])
        if i >= len(s2): return ord(s1[i])
        if s1[i] != s2[i]:
            return ord(s1[i]) - ord(s2[i])
    return 0
```

```python
def sc_strlen(s: str) -> int:
    count = 0
    for _ in s: count += 1
    return count

def sc_stricpy(dst: list, src: str, size: int) -> int:
    src_len = sc_strlen(src)
    if size > 0:
        copy_len = min(src_len, size - 1)
        for i in range(copy_len):
            if i < len(dst): dst[i] = src[i]
            else: dst.append(src[i])
        if len(dst) > copy_len: dst[copy_len] = '\0'
        else: dst.append('\0')
    return src_len

def sc_stricat(dst: list, src: str, size: int) -> int:
    dst_len = sc_strlen("".join(dst).split('\0')[0])
    src_len = sc_strlen(src)
    if size <= dst_len:
        return size + src_len

    copy_len = min(src_len, size - dst_len - 1)
    for i in range(copy_len):
        if dst_len + i < len(dst):
            dst[dst_len + i] = src[i]
        else:
            dst.append(src[i])

    if dst_len + copy_len < len(dst): dst[dst_len + copy_len] = '\0'
    else: dst.append('\0')

    return dst_len + src_len

def sc_strchr(s: str, c: str) -> str:
    for i in range(sc_strlen(s)):
        if s[i] == c:
            return s[i:]
    if c == '\0': return ""
    return None

def sc_strrchr(s: str, c: str) -> str:
    last_occurrence = None
    for i in range(sc_strlen(s)):
        if s[i] == c:
            last_occurrence = s[i:]
    if c == '\0': return ""
    return last_occurrence

def sc_strncmp(s1: str, s2: str, n: int) -> int:
    for i in range(min(n, sc_strlen(s1), sc_strlen(s2))):
        if s1[i] != s2[i]:
            return ord(s1[i]) - ord(s2[i])
    if n > sc_strlen(s1) and n <= sc_strlen(s2): return -ord(s2[sc_strlen(s1)])
    if n > sc_strlen(s2) and n <= sc_strlen(s1): return ord(s1[sc_strlen(s2)])
    return 0

def sc_strnstr(haystack: str, needle: str, length: int) -> str:
    if not needle: return haystack
    needle_len = sc_strlen(needle)
    for i in range(min(length, sc_strlen(haystack)) - needle_len + 1):
        match = True
        for j in range(needle_len):
            if haystack[i + j] != needle[j]:
                match = False
                break
        if match: return haystack[i:]
    return None

def sc_strdup(s: str) -> str:
    new_str = ""
    for char in s: new_str += char
    return new_str
```

```python
def sc_calloc(count: int, size: int) -> list:
    total_size = count * size
    buffer = ['\0'] * total_size
    return buffer
```

```python
def sc_substr(s: str, start: int, length: int) -> str:
    if not s: return None
    s_len = sc_strlen(s)
    if start >= s_len: return ""

    sub = ""
    for i in range(min(length, s_len - start)):
        sub += s[start + i]
    return sub

def sc_strjoin(s1: str, s2: str) -> str:
    if not s1 and not s2: return None
    if not s1: return sc_strdup(s2)
    if not s2: return sc_strdup(s1)

    joined = ""
    for char in s1: joined += char
    for char in s2: joined += char
    return joined

def sc_strtrim(s: str, set_chars: str) -> str:
    if not s: return None
    if not set_chars: return sc_strdup(s)

    start = 0
    end = sc_strlen(s) - 1

    while start <= end and sc_strchr(set_chars, s[start]):
        start += 1
    while end >= start and sc_strchr(set_chars, s[end]):
        end -= 1

    return sc_substr(s, start, end - start + 1)

def sc_split(s: str, c: str) -> list:
    if not s: return None
    result = []
    word = ""

    for char in s:
        if char == c:
            if word:
                result.append(word)
                word = ""
        else:
            word += char

    if word:
        result.append(word)

    return result

def sc_itoa(n: int) -> str:
    if n == 0: return "0"

    res = ""
    is_negative = n < 0

    if is_negative:
        n = -n

    while n > 0:
        res = chr((n % 10) + ord('0')) + res
        n //= 10

    if is_negative:
        res = "-" + res

    return res

def sc_strmapi(s: str, f) -> str:
    if not s or not f: return None
    res = ""
    for i in range(sc_strlen(s)):
        res += f(i, s[i])
    return res

def sc_striteri(s: list, f) -> None:
    if not s or not f: return
    for i in range(len(s)):
        s[i] = f(i, s[i])
```

```python
def sc_putchar_fd(c: str, fd: int) -> None:
    if sc_strlen(c) > 0:
        os.write(fd, c[0].encode('utf-8'))

def sc_putstr_fd(s: str, fd: int) -> None:
    if s:
        os.write(fd, s.encode('utf-8'))

def sc_putendl_fd(s: str, fd: int) -> None:
    if s:
        sc_putstr_fd(s, fd)
        sc_putchar_fd('\n', fd)

def sc_putnbr_fd(n: int, fd: int) -> None:
    num_str = sc_itoa(n)
    sc_putstr_fd(num_str, fd)
```
